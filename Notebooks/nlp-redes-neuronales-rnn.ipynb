# Introducción

Antes de empezar con las redes neuronales recurrentes (RNN), demos un repaso (a groso modo) del funcionamiento de las redes neuronales tradicionales:

1. Insertamos el dataset que vayamos a utilizar.
2. Se le aplicarán una serie de operaciones computacionales complejas.
3. Se predecirá un resultado.
4. Comparamos el resultado con el esperado, y de esto sacamos el error.
5. Este error se evalúa y se aplican nuevos pesos para mejorar la precisión.

Todos estos, pasos se podrán repetir hasta que consigamos un resultado que sea satisfactorio.

El principal problema de las tradicionales, es que no son capaces de procesar secuencias de datos, como videos o conversaciones (ya sea por texto o audio). Pero, que no sean capaces de procesar estas secuencias, no significan que no puedan, sino que no conocen la prioridad o el orden en la sencuencia. En definitiva, "saben lo que tienen pero no lo que tuvieron o tendrán". Por este motivo surgen las RNN, para procesar dichas secuencias de datos. Pero antes de ver las RNN, tendremos que conocer el **modelado de secuencia**.

### Modelado de secuencia (sequence modeling)

El modelado de secuencias es el proceso de predecir la siguiente palabra o carácter. Calcula la probabilidad de palabras que tendrán la oportunidad de ocurrir posteriormente en una secuencia particular. 

Este modelo tomará un valor de alta probabilidad de palabra o carácter como salida. A diferencia de ANN (Redes Neuronales Tradicionales), la salida del modelado de secuencia depende no solo de la entrada sino también de la salida anterior. Este es el proceso básico de la familia de RNNs.

### Redes Neuronales Recurrentes (RNN)

Es uno de los algoritmos más populares en el campo de la inteligencia artificial. Son capaces de superar a la mayoría de los algoritmos de la máquina en términos de velocidad computacional y precisión de alto rendimiento. 

Las redes neuronales se han mantenido firmes en asuntos de resolución de problemas complejos e investigación. Esta es una de las causas de que la inteligencia artificial se considere una innovación que cambia el mundo.

En términos simples, datos secuenciales utilizados para el mecanismo de búsqueda por voz detrás de dicha tecnología. 

Para implementar datos secuenciales de manera eficiente, el algoritmo responsable de convertirlos en una posibilidad son las redes neuronales recurrentes (RNN). 

Este es un algoritmo de última generación que puede memorizar la entrada debido a una memoria interna asociada con ella. Esto los hace ideales para problemas de aprendizaje automático que requieren datos secuenciales.

En este artículo explora el mundo de la inteligencia artificial y las RNNs. Que han sido algunos de los algoritmos notables que han sido fundamentales para lograr un tremendo éxito en el aprendizaje profundo en los últimos años.

# Redes Neuronales Recurrentes (RNN), ¿cómo funcionan?

Hasta ahora podríamos apreciar cierto nivel de inteligencia en el funcionamiento de las redes neuronales tradicionales. Son capaces de aprender a realizar tareas sencillas de la misma manera que lo haría un humano. Una ANN (como podría ser el caso de redes basadas en perceptrones) es capaz de establecer, por ejemplo, una clasificación de manera automática. Para realizar esta sencilla tarea, la red replica el *modus operandi* de un ser humano: aprender como se han clasificado ciertos objetos hasta la fecha (cuántos más se conozcan, mejor) y hallar los patrones comunes que comparten para saber cómo clasificar un objeto nuevo en función de a que tipo de objeto de los conocidos se parezca más.

Ahora bien, cuando se trata de predecir una trayectoria, un suceso, un evento... El humano es capaz de reconocer patrones que estas redes pasan por alto. 

Por poner un ejemplo: Nos ponemos en el caso de observar la trayectoria habitual de una persona. En nuestro caso, observamos diariamente la rutina de la persona y, al llegar a cierta calle con diversas entradas, vemos que elige una diferente cada día. Al cabo de un mes estudiando su comportamiento, vemos que: la mayoría de los días ha tomado la primera salida, casi todos los martes y jueves tomó la segunda, y, durante 3 días seguidos, tomó la tercera salida. 

<center><img src="https://imgur.com/CJWcVrE.png" width="700" height="700" /><img src="https://imgur.com/Vrah6FA.png" width="700" height="700" /></center>

Como humanos, podemos inferir que, si no es martes ni jueves, va a tomar con muy **alta probabilidad** la primera salida. Podemos **acordarnos** de la rutina que ha seguido e identificar que rutas son más **relevantes** para la persona. Sin realmente conocer la situación de la persona podemos hacernos a la idea de que, la manera **más óptima** de llegar a su destino habitual, es usar la primera salida y un suceso puntual le hizo cambiar de ruta por un breve periodo de tiempo. Además, la segunda salida es más importante para él ciertos días a la semana de manera regular. Teniendo toda esta información, no nos sería dificil predecir que salida va a coger la persona durante los siguientes días y, probablemente, tendríamos un alto porcentaje de acierto. Además si la persona cambiara de rutina, nos costaría un poco adaptarnos y aprender los cambios pero, lo **aprendido hasta ahora**, nos ayudaría a reconocer patrones antiguos en la nueva secuencia de sucesos.

Tras hablar con la persona, nos cuenta que a su casa se puede llegar por la primera y por la tercera salida. Nos cuenta que todos los martes y jueves trata de ir a entrenar a un parque cercano a su casa al que puede llegar a través de la segunda salida. También estuvo molesto hace tiempo porque cortaron la primera calle por obras durante 3 días y tuvo que seguir un camino más largo.

<center><img src="https://imgur.com/MtJNqtL.png" width="700" height="700" /></center>

Nosotros nos podíamos esperar esta respuesta sin sorprendernos demasiado. Sin embargo, una ANN no sería capaz de hacer el mismo razonamiento que nosotros y toda esta información sería nueva para ella. Esto se debe a que las ANN no tienen una **memoria** que les permita guardar la información de la secuencia para así establecer un vínculo entre los datos de entrada. Tampoco son capaces de solventar el problema de que las muestras de entrada no tengan un **tamaño fijo**.

### Memoria

Las RNN tratan de replicar el razonamiento ejemplificado anteriormente y aprender cosas nuevas sin olvidar las que ya se aprendieron. Este registro es lo que le permite a una red establecer un vínculo y una secuencia (con cierta probabilidad) sobre las muestras de entrada. 

La manera en la que se estructura la memoria de una RNN es la siguiente: Se estudian las capas ocultas (h) de la red neuronal en diferentes estados asociados a una unidad de tiempo (t). Para predecir una salida, además de la función de activación, se necesita la información de la capa en un estado anterior (h(t-1)).

# Tipos de Redes Neuronales Recurrentes (RNN)

Las RNN se dividen en varios tipos dependiendo de la cantidad de datos que tengamos como entrada o salida. Estos tipos se dividen en:
1. One-to-One (Uno a uno)
2. One-to-Many (Uno a muchos)
3. Many-to-One (Muchos a uno)
4. Many-to-Many (Muchos a muchos)

### One-to-One

Esto se conocía anteriormente como la RNN por defecto, generalmente caracterizado por una sola variable de entrada, como una palabra o imagen. Al mismo tiempo, las salidas se producen como un único valor de token. Todas las redes neuronales tradicionales entran en esta categoría. Incluso el tipo de tareas como **clasificación de spam** pueden entrar en esta categoría.

<center><img src="https://imgur.com/TuCrt4h.png" width="700" height="700" /></center>

### One-to-Many

Una única entrada usada para crear diversas salidas, un ejemplo de aplicación sería la **generación de música**.

<center><img src="https://imgur.com/rHuT4aE.png" width="700" height="700" /></center>

### Many-to-One

Consiste en diversas variables de entrada que generan una única salida, por ejemplo el **análisis de sentimientos**. Esto permite identificar un sentimiento a partir de un grupo de palabras, determinar la palabra que falta para terminar la frase recibida en entrada. Para aprender más sobre el análisis de sentimientos, consulta este artículo.

<center><img src="https://imgur.com/Jjw6qGY.png" width="700" height="700" /></center>

### Many-to-Many

Se utilizan varias entradas para generar varias salidas. El **reconocimiento de nombres de entidades (NER)** es una aplicación famosa que utiliza este tipo de RNN o la **traducción de textos**.

<center><img src="https://imgur.com/wOs7ry3.png" width="700" height="700" /></center>

Una vez que hemos visto los tipos de RNN, tendremos que conocer como funciona la **retropropagación**, para reducir la pérdida y optimizar los pesos asignados. El entrenamiento de una RNN es muy similar a cualquier otra red neuronal tradicional. El uso de un algoritmo de retropropagación ha sido una gran adición al procedimiento de entrenamiento. El objetivo de usar retropropagación es volver a través de la red neuronal de tal manera que cualquier derivada parcial del error se identifique con respecto a los pesos. Esto nos permite eliminar tales valores de los pesos. Las derivadas se utilizan por el descenso de gradiente para minimizar una función de pérdida. Los pesos se ajustan según la forma en que pueden disminuir las tasas de error. 


# Redes Neuronales Recurrentes (RNN), ¿tienen memoria de pez?

Imaginemos la siguiente frase:
- Me gustan los mochis, me voy a comer a ...


La RNN no es capaz de retener la palabra mochis para predecir Japón. Para llegar a determinar Japón, es necesario que la RNN posea una memoria mejor. Para ello hay que conseguir que las neuronas sean más complejas. En particular, vamos a ver el caso de la LSTM (Long Short Term Memory). Además del estado escondido convencional h_t, vamos a añadir un segundo estado c_t. Aquí, h_t representa la memoria corta neuronal y c_t la memoria larga.

No vamos a entrar en las consideraciones técnicas, lo esencial es observar que tenemos una célula mucho más compleja, que nos permite solucionar el problema de memoria. Con la LSTM, vamos a pasar las h_t y c_t por puertas, 4 en total.

1. La primera puerta permite eliminar la información inútil, es la forget gate.
2. La segunda puerta, almacena la información nueva, la store gate.
3. La tercera puerta actualiza la información que vamos a dar a la RNN con el resultado de la forget gate y la store gate, es la update gate.
4. Y la última puerta (output gate), nos da y_t y h_t.

Este largo proceso permite controlar la información que conservamos y transmitimos a través del tiempo.

La RNN llega a saber qué hay que guardar y qué hay que olvidar, gracias a su aprendizaje. La LSTM (Long Short Term Memory) no es única, también podemos usar GRU (Gated Recurrent Unit), solo cambia la arquitectura de la célula.
