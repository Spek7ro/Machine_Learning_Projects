### Métricas para Regresión 

Recurso: https://medium.com/@nicolasarrioja/m%C3%A9tricas-en-regresi%C3%B3n-5e5d4259430b

- Error: El error es un concepto muy sencillo, simplemente es la diferencia que existe entre el valor que nuestro modelo ha predicho y el valor real de la observación con la que estamos haciendo el test.



<h3 style="color:red">Error absoluto máximo (M)</h3>

$y$ : valor verdadero 

$\hat{y}$ : valor predicho 

$$\text{M}(y, \hat{y}) = max(\left| y_i - \hat{y}_i \right|)$$

- Es la desviación máxima entre el valor real y la medición.

```python
from sklearn.metrics import max_error

y_verdadero = [1, 2, 3, 4, 5]
y_predicho = [1, 2, 3, 4, -5]
max_error(y_verdadero, y_predicho)
```

<h3 style="color:red">Error absoluto medio (mean absolute error - MAE)</h3>

$y$ : valor verdadero 

$\hat{y}$ : valor predicho 

$n$ : tamaño de la muestra 

$$\text{MAE}(y, \hat{y}) = \frac{1}{n} \sum_{i=1}^{n} \left| y_i - \hat{y}_i \right|$$

```python
from sklearn.metrics import mean_absolute_error

y_verdadero = [1, 2, 3, 4, 5]
y_predicho = [1, 2, 3, 4, -5]
mean_absolute_error(y_verdadero, y_predicho)
```

<h3 style="color:red">Error cuadrático medio (mean squared error - MSE)</h3>

$y$ : valor verdadero 

$\hat{y}$ : valor predicho 

$n$ : tamaño de la muestra 

$$\text{MSE}(y, \hat{y}) = \frac{1}{n} \sum_{i=1}^{n} (y_i - \hat{y}_i)^2$$

```python
from sklearn.metrics import mean_squared_error

y_verdadero = [1, 2, 3, 4, 5]
y_predicho = [1, 2, 3, 4, -5]
mean_squared_error(y_verdadero, y_predicho)
```

<h3 style="color:red">Suma de los cuadrados de los residuos (RSS)</h3>

$y$ : valor verdadero 

$\hat{y}$ : valor predicho 

$n$ : tamaño de la muestra 

$$\text{RSS}(y, \hat{y}) = \sum_{i=1}^{n} (y_i - \hat{y}_i)^2$$

```python
from sklearn.metrics import mean_squared_error

y_verdadero = [1, 2, 3, 4, 5]
y_predicho = [1, 2, 3, 4, -5]
mean_squared_error(y_verdadero, y_predicho)*len(y_predicho)
```

<h3 style="color:red">Raíz cuadrada del error cuadrático medio (RMSE)</h3>

$y$ : valor verdadero 

$\hat{y}$ : valor predicho 

$n$ : tamaño de la muestra 

$$\text{RMSE}(y, \hat{y}) = \sqrt{\frac{1}{n} \sum_{i=1}^{n} (y_i - \hat{y}_i)^2}$$

```python
from sklearn.metrics import mean_squared_error

y_verdadero = [1, 2, 3, 4, 5]
y_predicho = [1, 2, 3, 4, -5]
mean_squared_error(y_verdadero, y_predicho)
```

<h3 style="color:red">R^2 (Coeficiente de determinación)</h3>

$y$ : valor verdadero 

$\hat{y}$ : valor predicho 

$\bar{y}$ : promedio de los valores verdaderos

$n$ : tamaño de la muestra 

$$R^2(y, \hat{y}) = 1 - \frac{\sum_{i=1}^{n} (y_i - \hat{y}_i)^2}{\sum_{i=1}^{n} (y_i - \bar{y})^2}$$

```python
from sklearn.metrics import r2_score

y_verdadero = [1, 2, 3, 4, 5]
y_predicho = [1, 2, 3, 4, 2]
r2_score(y_verdadero, y_predicho)
```
