```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
```

```python
m = 100
X = 6 * np.random.rand(m, 1) - 3
# print(X)
y = 2 + X + 0.5 * X ** 2 + np.random.rand(m, 1) ## 2x + 0.5x^2
```

```python
plt.scatter(X, y)
plt.show()
```

```python
poly_features = PolynomialFeatures(degree=2, include_bias=False) ## Crear un polinomio de grado 2
X_ploy = poly_features.fit_transform(X)

print(X[0], X_ploy[0]) ## Conseva el primer elemento de X y X_ploy se eleva al cuadrado
```

```python
## Entrenar el modelo de regresión lineal con las características polinomiales
lin_reg = LinearRegression()
lin_reg.fit(X_ploy, y)

print('Intercepción:', lin_reg.intercept_)
print('Coeficientes:', lin_reg.coef_)
print(f"{lin_reg.intercept_[0]:.2f} + {lin_reg.coef_[0][0]:.2f} * x + {lin_reg.coef_[0][1]:.2f} * x^2")
```

```python
X_plot = np.linspace(-3, 3, m).reshape(m, 1)
X_plot_poly = poly_features.transform(X_plot)
y_plot = lin_reg.predict(X_plot_poly)

plt.figure(figsize=(8, 6))
plt.scatter(X, y, color='black', label='Datos', s=20)
plt.plot(X_plot, y_plot, 'r-', label='Regresión lineal')
plt.legend()
plt.show()
```
