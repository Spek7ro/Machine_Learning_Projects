```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
U = lambda T: 204165.5/(330 - 2*T) + 10400/(T-20) # funcion
dU = lambda T, dt: (U(T + dt) - U(T)) / dt  # derivada
```

```python
# Gradica de la funcion
t = np.linspace(40, 90, 100)
# plt.plot(t, U(t))
plt.plot(t, dU(t, 0.01))
plt.grid(True)
```

```python
ti = 40
ti_optimo = 40
minimo = np.inf
t_i=np.arange(40,90, 0.001)

for ti in t_i:
    # Aproximacion de la derivada
    if U(ti) < minimo:
        minimo = U(ti)
        ti_optimo = ti

print(f"Minimo en {ti} es {minimo}")
print(f"Minimo en {ti_optimo} es {minimo}")
```

```python
def biseccion(f, a, b, tol=0.0001, max_iter=20):
    if f(a) * f(b) >= 0:
        print("No se pude realizar ")
        return None

    for i in range(max_iter):
        c = (a + b) / 2  
        valor_f = f(c)

        if abs(valor_f) < tol or (b - a) / 2 < tol:
            return c
        
        if np.sign(valor_f) == np.sign(f(a)):
            a = c
        else:
            b = c
        print(f"Iteracion {i}: a = {a}, b = {b}, c = {c}, f(c) = {valor_f}")  
    
    return (a + b) / 2

t_min = biseccion(U, 40, 90)
print(t_min, U(t_min))
```
