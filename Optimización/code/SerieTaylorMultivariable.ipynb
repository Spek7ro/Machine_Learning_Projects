```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib.lines import Line2D
```

```python
### Ejemplo 2: e^x+y

f_orig = lambda x, y: np.exp(x + y)
p1 = lambda x, y: 1 + x + y
p2 = lambda x, y: 1 + x + y + 0.5 * (x**2 + 2*x*y + y**2)

r = np.linspace(-1, 1, 40)
X, Y = np.meshgrid(r, r)

Z_orig = f_orig(X, Y)
Z_p1 = p1(X, Y)
Z_p2 = p2(X, Y)

### Grafica en 3D
fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(111, projection='3d')

## Superficie original
ax.plot_surface(X, Y, Z_orig, cmap='Blues', alpha=0.4, label='Original')

### crear palano de 1er Orden 
ax.plot_wireframe(X, Y, Z_p1, color='red', alpha=0.5, rstride=5, cstride=5, label='1er Orden')

### crear plano de 2do Orden
ax.plot_wireframe(X, Y, Z_p2, color='green', alpha=0.7, rstride=4, cstride=4, label='2do Orden')

ax.scatter([0], [0], [1], color='black', s=100)
ax.set_title(r'Taylor: $f(x,y) = e^{x+y}$ en (0,0)', fontsize=14)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

legend_el = [
    Line2D([0], [0], color='blue', lw=4, alpha=0.4, label=r'Original $e^{x+y}$'),
    Line2D([0], [0], color='red', lw=1, label=r'1er Orden: $1+x+y$'),
    Line2D([0], [0], color='green', lw=1, label=r'2do Orden: $1+(x+y)+\frac{(x+y)^2}{2}$')
]
ax.legend(handles=legend_el)

plt.show()
```
