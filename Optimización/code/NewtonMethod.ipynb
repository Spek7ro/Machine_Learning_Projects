```python
f = lambda x: 204165.5/(330-2*x) + 10400/(x-20)
fdx = lambda x,dx: (f(x+dx) - f(x-dx))/(2*dx)
fddx = lambda x, dx: (f(x+dx)-2*f(x)+f(x-dx))/(dx**2)
```

```python
x0 = 90
epsilon = 0.0001
Xn  = 90.0
dx = 0.001

for i in range(100):
        fx = f(Xn)
        Xn2 = Xn -(fdx(Xn,dx) / fddx(Xn,dx))

        if(abs(Xn2 - Xn) < epsilon):
            print("Fin num de interaciones:", i + 1)
            print("Nuevo punto:", Xn2)
            print("Valor de la funcion:", fx)
            break

        Xn = Xn2
```
