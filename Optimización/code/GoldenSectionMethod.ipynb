```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
U = lambda T: 204165.5/(330 - 2*T) + 10400/(T-20) # funcion
dU = lambda T, dt: (U(T + dt) - U(T)) / dt  # derivada
```

```python
t = np.linspace(40, 90, 100)
# plt.plot(t, U(t))
plt.plot(t, dU(t, 0.01))
plt.grid(True)
```

```python
a = 40
b = 90
tau = 0.618 # Proporcion aurea
epsilon = 0.000001 # Tolerancia

for i in range(100):
    x1 = a*tau+b*(1-tau)
    x2 = a*(1-tau)+b*tau
    if U(x1) < U(x2):
        b = x2
    else:
        a = x1
    if abs(U(x1) - U(x2)) < epsilon:
        break

x = (a + b) / 2
print("Optimo x: ", x)
print("U(x): " , U(x))
# print("U'(x):" , dU(x, 0.01))
print("iteraciones: ", i)
```
