```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x_var = np.arange(0, 1, 0.01)  # variable aleatoria X € [0,1]

# funcion original:  2sen(5x)+3x^3-2x^2+3x-5
f  = lambda  x: 2*np.sin(5*x)+3*x**3-2*x**2+3*x-5

# Derivada analitica: primera derivada
df_analitica = lambda x: 10*np.cos(5*x)+9*x**2-4*x+3 
```

```python
def derivada_adelantada(f, x, h):
    return (f(x + h) - f(x)) / h

# derivada_adelantada = lambda x, h: (f(x + h) - f(x)) / h

def derivada_atrasada(f, x, h):
    return (f(x) - f(x - h)) / h

def derivada_centrada(f, x, h):
    return (f(x + h) - f(x - h)) / (2 * h)
```

```python
h = 0.01 # incremento 
    
analitica = df_analitica(x_var)
adelantada = derivada_adelantada(f, x_var, h)
atrasada = derivada_atrasada(f, x_var, h)   
centrada = derivada_centrada(f, x_var, h)
```

```python
### Grafica de las derivadas
plt.figure(figsize=(12, 6))

# Grafica derivada analitica exacta
plt.plot(x_var, analitica, 'k-', linewidth=2, alpha=0.8, label='Exacta (Analítica)')

# Graficas de las derivadas aproximacimadas
plt.plot(x_var, adelantada, 'g--', label=f'Adelantada (h={h})')
plt.plot(x_var, atrasada, 'r--', label=f'Atrasada (h={h})')
plt.plot(x_var, centrada, 'b.', markersize=4, label=f'Centrada (h={h})')

plt.title(r"Derivada de $f(x) = 2\sin(5x)+3x^3-2x^2+3x-5$")
plt.xlabel("x")
plt.ylabel("f'(x)")
plt.legend()
plt.grid(True, linestyle='--', alpha=0.6)

plt.show()
```
