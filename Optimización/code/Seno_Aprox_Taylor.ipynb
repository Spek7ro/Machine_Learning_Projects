```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
def taylor_sin(x, n_terminos):
    y = np.zeros_like(x)
    for n in range(n_terminos):
        coeficiente = (-1)**n
        exponente = 2*n + 1
        termino = (coeficiente * (x**exponente)) / math.factorial(exponente)
        y += termino
    return y

x = np.linspace(-2 * np.pi, 2 * np.pi, 500)
y_real = np.sin(x)

# gráfica
plt.figure(figsize=(10, 6))
plt.plot(x, y_real, label='sin(x) Real', color='black', linewidth=2)

# 3. Graficar aproximaciones con diferentes números de términos
for n in [1, 3, 5, 7, 10]:
    y_aprox = taylor_sin(x, n)
    plt.plot(x, y_aprox, linestyle='--', label=f'Taylor (n={n} términos)')

# 4. Estética de la gráfica
plt.ylim(-1.5, 1.5)
plt.axhline(0, color='gray', lw=1)
plt.axvline(0, color='gray', lw=1)
plt.title('Aproximación de la Serie de Taylor para sin(x)')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```
