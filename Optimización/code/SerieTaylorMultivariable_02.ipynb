```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
from matplotlib.lines import Line2D
```

```python
# función original y sus aproximaciones de orden 1 y 2
f_real = lambda x1, x2: 3*x2 - (x1 / x2)
# Aproximación de 1er orden:
f_primer_orden = lambda x1, x2: -x1 + 5*x2 - 2
# Aproximación de 2do orden: Términos lineales + Términos cuadráticos
f_segundo_orden= lambda x1, x2: f_primer_orden(x1, x2) + (x1 - 2)*(x2 - 1) - 2*(x2 - 1)**2
```

```python
x1_vals = np.linspace(1, 3, 30)
x2_vals = np.linspace(0.5, 1.5, 40)
X1, X2 = np.meshgrid(x1_vals, x2_vals)

fig = plt.figure(figsize=(15, 9))
ax = fig.add_subplot(111, projection='3d')
surf1 = ax.plot_surface(X1, X2, f_real(X1, X2), cmap='Blues', alpha=0.3)
# Primer Orden 
surf2 = ax.plot_wireframe(X1, X2, f_primer_orden(X1, X2), color='red', alpha=0.4, linewidth=0.8)
# Segundo Orden
surf3 = ax.plot_surface(X1, X2, f_segundo_orden(X1, X2), cmap='YlOrRd', alpha=0.5)

ax.scatter(2, 1, 1, color='black', s=100, label='Punto (2,1)')

# Estética y etiquetas
ax.set_title('Aproximación de Taylor Multivariable $f(X_1, X_2) = 3 X_2 - \\frac{X_1}{X_2}$', fontsize=16)
ax.set_xlabel('$X_1$')
ax.set_ylabel('$X_2$')
ax.set_zlabel('$f(X_1, X_2)$')

legend_elements = [
    Line2D([0], [0], color='blue', lw=4, label='Función Real', alpha=0.3),
    Line2D([0], [0], color='red', lw=1, label='1er Orden malla roja'),
    Line2D([0], [0], color='orange', lw=4, label='2do Orden malla naranja'),
    Line2D([0], [0], marker='o', color='w', markerfacecolor='black', markersize=10, label='Punto (2,1)')
]
ax.legend(handles=legend_elements, loc='upper left')

plt.show()
```
