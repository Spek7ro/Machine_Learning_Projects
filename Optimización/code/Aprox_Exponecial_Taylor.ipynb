```python
import numpy as np
import matplotlib.pyplot as plt
from math import factorial

def taylor_exp(x, num_terms):
    return sum((x**n) / factorial(n) for n in range(num_terms))

x = np.linspace(-3, 3, 100) 
y_real = np.exp(x)          

plt.figure(figsize=(10, 6))

# función real
plt.plot(x, y_real, label="e^x (Real)", color="black", linewidth=2, linestyle="--")

# Dibujar diferentes aproximaciones con distintos números de términos
terminos = [1, 2, 3, 5, 10]
for n in terminos:
    y_approx = [taylor_exp(val, n) for val in x]
    plt.plot(x, y_approx, label=f"Taylor (n={n})")

plt.title("Aproximación de e^x mediante Series de Taylor")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.ylim(-1, 20) # Limitamos el eje Y para mejor visibilidad
plt.legend()
plt.grid(True, linestyle=':', alpha=0.6)

# Mostrar resultado
plt.show()
```
