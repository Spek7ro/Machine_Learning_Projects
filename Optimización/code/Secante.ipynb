```python
import numpy as np
```

```python
U = lambda T: 204165.5/(330 - 2*T) + 10400/(T-20) # funcion
dU = lambda x, delta: (U(x+delta)-U(x))/delta
```

```python
delta = 0.001
epsilon = delta
x1 = 90
x2 = 40

for i in range(1,100):
    alpha = x2 - ((dU(x2,delta)*(x2-x1)))/(dU(x2,delta)-dU(x1,delta))
    if U(alpha) * dU(x1,delta) < 0:
        x2 = alpha
    else:
        x1 = alpha
    if abs(dU(alpha,delta)) < epsilon:
        break

print("x: %f" % alpha)
print("fx: %f" % U(alpha))
print("f'x: %f" % dU(alpha,delta))
print("iteraciones: %d" % i)
```
