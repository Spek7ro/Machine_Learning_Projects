```python
import pandas as pd
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
df = pd.read_csv('../Datasets/machine_failure_dataset.csv')
df = pd.get_dummies(df, columns=['Machine_Type'])
df.head()
```

```python
X = df.drop('Failure_Risk', axis=1).values.astype(float)
y = df['Failure_Risk'].values
```

```python
m = np.mean(X, axis=0)
std = np.std(X, axis=0)
X = (X - m) / (std + 1e-8)
```

```python
# Modelo MLP
model = Sequential([
    # Capa oculta: 8 neuronas, activaciГіn ReLU
    Dense(8, activation='relu', input_shape=(X.shape[1],)),
    # Capa de salida: 1 neurona (binaria), activaciГіn Sigmoide
    Dense(1, activation='sigmoid')
])

model.compile(optimizer='adam', 
              loss='binary_crossentropy', 
              metrics=['accuracy'])
model.summary()
```

```python
model.fit(X, y, epochs=50, batch_size=32, verbose=1)
```

```python
loss, accuracy = model.evaluate(X, y, verbose=0)
print(f"PrecisiГіn (Accuracy): {accuracy * 100:.2f}%")
print(f"Loss: {loss:.4f}")
```
