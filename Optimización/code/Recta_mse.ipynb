```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# cargar el dataset
df = pd.read_csv('../Datasets/Salary_Data.csv')
```

```python
df.describe().T
```

```python
# Grafica salario vs años de experiencia del dataset
plt.plot(df.iloc[:,0], df.iloc[:,1], color='red', marker='o', markerfacecolor='blue', markersize=7)
plt.xlabel(df.columns[0])
plt.ylabel(df.columns[1])
plt.show()
```

```python
x = df.iloc[:, 0].values 
y = df.iloc[:, 1].values

a = np.sum(x*y) / np.sum(x**2)
x_p = np.array([np.min(x), np.max(x)])
y_p = a * x_p 

plt.plot(x_p, y_p, color='black') 
plt.plot(x, y, 'o', color='red')
# for i in range(len(x)):
#     plt.plot([x[i], x[i]], [y[i], a*x[i]], color='blue')
plt.xlabel(df.columns[0], fontsize=14)
plt.ylabel(df.columns[1], rotation=90, fontsize=14)
plt.show()
```

```python
N = df.shape[0] # numero de datos
suma_xy = np.sum(x*y)
suma_y = np.sum(y)
suma_x = np.sum(x)
suma_x2 = suma_x**2
suma_xx = np.sum(x**2)

a = (N*suma_xy - suma_y*suma_x) / (N*suma_xx- suma_x2) # pendiente
# b = (suma_y*suma_xx-suma_xy*suma_x) / (N*suma_xx- suma_x2) # intercepto
b2 = (suma_y - a * suma_x) / N # intercepto

print('pendiente a =', a)
print('intercepto b =', b2)

# plotear la recta 
x_p = np.array([np.min(x), np.max(x)])
y_p = a * x_p + b2
plt.plot(x_p, y_p,'r--')
plt.plot(x, y, 'bo')
for i in range(len(x)):
    plt.plot([x[i], x[i]], [y[i], a*x[i]+b2], 'g-', linewidth=0.8)
plt.xlabel(df.columns[0], fontsize=15);
plt.ylabel(df.columns[1], fontsize=15);
plt.title('modelo: y=%.2f x+ %.2f'% (a, b2), fontsize=12);
```
